### 3. Identifying and Removing Inconsistencies


<h5>3.1 Identify inconsistent or irrelevant entries in specific columns (e.g., Country).</h5>


```python
# Write your code here
group_country = df.groupby('Country')['Country'].value_counts()
group_country_df = group_country.reset_index(name='Counts')
print(group_country_df.to_string())
```

```python
group_country_df.to_csv("country.csv")
```

<h5>3.2 Standardize entries in columns like Country or EdLevel by mapping inconsistent values to a consistent format.</h5>


```python
## Write your code here
df_country_std = pd.read_csv('standardize-country.csv', encoding='latin1')
df_country_std
```

```python
for index, row in df_country_std.iterrows():
    country = row['Country']
    std_country = row['standardized_country']
    df['Country'].replace(country, std_country, inplace=True)
```

```python
#recheck the data
group_country = df.groupby('Country')['Country'].value_counts()
group_country_df = group_country.reset_index(name='Counts')
print(group_country_df.to_string())
```

### 4. Encoding Categorical Variables


<h5>4.1 Encode the Employment column using one-hot encoding.</h5>


```python
## Write your code here
# Split the 'Employment' column into lists
df['Employment_List'] = df['Employment'].str.split(';')
df = df.explode('Employment_List')
df_one_hot = pd.get_dummies(df['Employment_List'], prefix='Employment')
df = pd.concat([df, df_one_hot], axis=1)
df.drop("Employment", axis = 1, inplace=True)
df
```

```python
df.drop("Employment_List", axis = 1, inplace=True)
df
```

### 5. Handling Missing Values


<h5>5.1 Identify columns with the highest number of missing values.</h5>


```python
## Write your code here
missing_values_count = df.isnull().sum()
sorted_missing_values = missing_values_count.sort_values(ascending=False)
print(sorted_missing_values)
```

<h5>5.2 Impute missing values in numerical columns (e.g., `ConvertedCompYearly`) with the mean or median.</h5>


```python
## Write your code here
import numpy as np
mean_CCY = df['ConvertedCompYearly'].mean()
df['ConvertedCompYearly'].replace(np.nan, mean_CCY, inplace=True)

df['ConvertedCompYearly'].isnull().value_counts()
```

<h5>5.3 Impute missing values in categorical columns (e.g., `RemoteWork`) with the most frequent value.</h5>


```python
## Write your code here
most_RW = df.groupby ('RemoteWork')['RemoteWork'].value_counts().idxmax()
df['RemoteWork'].replace(np.nan, most_RW, inplace=True)

df['RemoteWork'].isnull().value_counts()
```

### 6. Feature Scaling and Transformation


<h5>6.1 Apply Min-Max Scaling to normalize the `ConvertedCompYearly` column.</h5>


```python
## Write your code here
X_min = df['ConvertedCompYearly'].min()
X_max = df['ConvertedCompYearly'].max()

df['ConvertedCompYearly_MinMax'] = (df['ConvertedCompYearly'] - X_min) / (X_max - X_min)
df['ConvertedCompYearly_MinMax']
```

<h5>6.2 Log-transform the ConvertedCompYearly column to reduce skewness.</h5>


```python
## Write your code here
df['ConvertedCompYearly_Log'] = np.log(df['ConvertedCompYearly'] + 1)
print(df[['ConvertedCompYearly', 'ConvertedCompYearly_Log']].head())
```

### 7. Feature Engineering


<h5>7.1 Create a new column `ExperienceLevel` based on the `YearsCodePro` column:</h5>


```python
## Write your code here
df['YearsCodePro'].replace(np.nan, 0, inplace=True)
df['YearsCodePro'].replace('Less than 1 year', 0, inplace=True)
df['YearsCodePro'].replace('More than 50 years', 51, inplace=True)
df['YearsCodePro'] = df['YearsCodePro'].astype(int)
```

```python
exp_level = ['Entry', 'Intermediate', 'Senior']
bins = [0, 5, 10, 51]
df['ExperienceLevel'] = pd.cut(df['YearsCodePro'], bins, labels=exp_level, include_lowest=True )
df[['YearsCodePro','ExperienceLevel']].head(20)
```
